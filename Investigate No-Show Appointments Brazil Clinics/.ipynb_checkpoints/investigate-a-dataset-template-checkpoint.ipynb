# Cleaning Data

```python
# We will not use the PatientId, AppointmentID, ScheduledDay, and AppointmentDay. So, We will start cleaning by dropping those coloumns
df.drop(['PatientId','AppointmentID','ScheduledDay','AppointmentDay'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.rename(columns= lambda x: x.replace('-','_'))
```

```python
# It was found that Handcap coloumn states either patient is handcapped or not (0 or 1), but it was found that there is a values of 2, 3, and 4 so we had to remove them.
i = 2
for i in range (2,5):
    index_names = df[df['Handcap'] == i].index
    df.drop(index_names, inplace=True)
```

```python
# Verify results
df['Handcap'].unique()
```

## Exploratory Data Analysis

### General Data Showing

```python
df.hist(figsize=[10,10]);
```

* Most of patients doesnot suffer from Diabetes.
* Most of Patients aren't Handicaps.
* Around 18% of the patients (20000 out of 110000) suffer from hipertension (High Blood Pressure).
* Most of the patients aren't enrolled in the brazillian welfare program (about 9% only enrolled).
* Only Half of the patients recevied a SMS to confirm their appointment

```python
# Assign Names to patients who showed and doesn't showed to make the recalling and comparison easier
Show = df['No-show'] == 'No'
No_Show = df['No-show'] == 'Yes'
```

### Research Question 1  (Descovering If Age influence showing for the appointment)

```python
Age_show = df['Age'][Show]
Age_no_show = df['Age'][No_Show]
plt.figure(figsize=[10,10])
plt.hist(Age_show,alpha=0.5,label='Show')
plt.hist(Age_no_show,alpha=0.5,label='No_Show',color='Red')
plt.legend()
plt.title('Showing & No-Showing for the appointments according to Age')
plt.xlabel('Age')
plt.ylabel('Patients Number');
```

* Patients with age ranges between 0 - 15 years old are most likely to attend there appointments more than any other age range follwed by people with age range 45 - 60.
* As patients age pass 60 they are most likely to skip there appointments.

### Research Question 2  (Descovering If Gender influence showing for the appointment)

```python
Gender_show = df['Gender'][Show]
Gender_no_show = df['Gender'][No_Show]
plt.figure(figsize=[7.50, 7.50])
plt.hist(Gender_show,alpha=0.5,label='Show',color='Blue',edgecolor='black',)
plt.hist(Gender_no_show,alpha=0.5,label='No_Show',color='Red',edgecolor='black')
plt.legend()
plt.title('Showing & No-Showing for the appointments according to gender')
plt.xlabel('Gender')
plt.ylabel('Patients Number');
```

```python
print(df['Gender'][Show].value_counts())
print(df['Gender'][No_Show].value_counts())
```

* Gender isn't significant to compare by it (Females Showed for appointments are more than males and females who didn't show is more thant males).

### Research Question 3  (Descovering If enrolling in Brazilian Welfare Program Bolsa Family influence showing for the appointment)

```python
Scholarsip_show = df['Scholarship'][Show]
Scholarsip_no_show = df['Scholarship'][No_Show]
plt.figure(figsize=[10,10])
plt.hist(Scholarsip_show,alpha=0.5,label='Show')
plt.hist(Scholarsip_no_show,alpha=0.5,label='No_Show',color='Red')
plt.legend()
plt.title('Showing & No-Showing for the appointments according to Scholarship')
plt.xlabel('Scholarship Enrollment')
plt.ylabel('Patients Number');
```

```python
print(df['Scholarship'][Show].value_counts())
print(df['Scholarship'][No_Show].value_counts())
```

* Being enrolled in bolsa family welfare program isn't significant to compare by it (Number of bolsa enrolled who showed is less than the un-enrolled who showed but also number of bolsa enrolled who didn't show is more than the un-enrolled who didn't show)..

### Research Question 7  (Descovering If Receiving Reminding SMS influence showing for the appointment)

```python
SMS_show = df['SMS_received'][Show]
SMS_no_show = df['SMS_received'][No_Show]
plt.figure(figsize=[10,10])
plt.hist(SMS_show,alpha=0.5,label='Show')
plt.hist(SMS_no_show,alpha=0.5,label='No_Show',color='Red')
plt.legend()
plt.title('Showing & No-Showing for the appointments according to Receiving Reminding SMS')
plt.xlabel('SMS Received')
plt.ylabel('Patients Number');
```

### Research Question 8  (Descovering If Neighbourhood where patients live at influence showing for the appointment)

```python
Neighbourhood_show = df['Neighbourhood'][Show]
Neighbourhood_no_show = df['Neighbourhood'][No_Show]
plt.figure(figsize=[15,15])
Neighbourhood_show.value_counts().plot(kind = 'bar',label='Show',color='Blue',alpha = 0.7)
Neighbourhood_no_show.value_counts().plot(kind = 'bar',label='No_Show',color='Red',alpha = 0.75)
plt.legend()
plt.title('Showing & No-Showing for the appointments according to Neighbourhood')
plt.xlabel('Neighbourhood')
plt.ylabel('Patients Number');
```

* It is very obvious that the Neighbourhood affects the number of patients shows for the appointments at the clinic, To Know why we must gather data about those neighbourhoods and answer some questions.
